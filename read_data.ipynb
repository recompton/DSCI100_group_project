```r
library(tidyverse)
library(lubridate)
air_quality <- read_csv("AirQualityUCI.csv") |>
#    select(-"...16", -"...17") |>
#    filter("CO(GT)" != -200.0, "PT08.S1(CO)" != -200, "NMHC(GT)" != -200, "PT08.S2(NMHC)" != -200, "NOx(GT)" != -200, "PT08.S3(NOx)" != -200, "NO2(GT)" != -200, "PT08.S4(NO2)" != -200, "PT08.S5(O3)" != -200, na.rm = TRUE) |>
    slice(1:9357) |>
    mutate(Year = year(mdy(Date))) |>
    mutate(Month = month(mdy(Date))) |>
    mutate(Day = day(mdy(Date))) |>
    select(Time, PT08.S1_CO, T, RH, AH, Year, Month, Day) |>
    filter(PT08.S1_CO != -200)
air_quality
```

```r
air_quality_monthly_mean <- air_quality |>
    group_by(Month, Year) |>
    summarize(monthly_mean = mean(PT08.S1_CO))
air_quality_monthly_mean

air_quality_monthly_t <- air_quality |>
    group_by(Month, Year) |>
    summarize(monthly_t = mean(T))
air_quality_monthly_t

air_quality_monthly_rh <- air_quality |>
    group_by(Month, Year) |>
    summarize(monthly_rh = mean(RH))
air_quality_monthly_rh

air_quality_monthly_ah <- air_quality |>
    group_by(Month, Year) |>
    summarize(monthly_ah = mean(AH))
air_quality_monthly_ah

air_quality_mean <- list(air_quality_monthly_mean, air_quality_monthly_t, air_quality_monthly_rh, air_quality_monthly_ah) |>
    reduce(full_join, by = c("Year", "Month"))
air_quality_mean
```

```r
air_quality_plot <- ggplot(air_quality, aes(x = Month, y = PT08.S1_CO)) +
    geom_point()
air_quality_plot
```
